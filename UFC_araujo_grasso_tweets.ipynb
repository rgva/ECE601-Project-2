```python
import tweepy
import webbrowser
import time
import pandas as pd
from textblob import TextBlob
#Authentication script
consumer_key = ""
consumer_secret = ""
callback_uri = 'oob'
auth = tweepy.OAuthHandler(consumer_key, consumer_secret, callback_uri)
redirect_url = auth.get_authorization_url()
print(redirect_url)
webbrowser.open(redirect_url)
user_pint_input = input("What is the pin value? ") #User pin input
auth.get_access_token(user_pint_input)
print(auth.access_token, auth.access_token_secret) 
api = tweepy.API(auth)
```

```python
#Get tweets related to UFC Fight Night: Grasso v. Araujo (Oct. 15th 2022)
def get_fight_tweets(key_word):

    twitter_usernames = []
    tweet_time = []
    tweet_string = []

    for tweet in tweepy.Cursor(api.search_tweets,q=key_word,count=1000).items(10000):
        #No retweets
        if (not tweet.retweeted) and ('RT @' not in tweet.text):
        #Only English
            if tweet.lang == "en":
                twitter_usernames.append(tweet.user.name)
                #Date and Time
                tweet_time.append(tweet.created_at)
                #Tweet text
                tweet_string.append(tweet.text)
    print([tweet.user.name,tweet.created_at,tweet.text])
#Pandas dataframe
    df = pd.DataFrame({'name':twitter_usernames, 'time': tweet_time, 'tweet': tweet_string})

    return df
```

```python
get_fight_tweets('Alexa Grasso')
```

```python
get_fight_tweets('Viviane Araujo')
```

```python
def sentiment_polarity(key_word):
    tweet_string = []
    for tweet in tweepy.Cursor(api.search_tweets,q=key_word,count=1000).items(10000):
        analyze = TextBlob(str(tweet))
    tweet_string.append(tweet.text)

    if analyze.sentiment.polarity > 0:
        return 'positive'
    elif analyze.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'
```

```python
print(sentiment_polarity('Viviane Araujo'))
```
